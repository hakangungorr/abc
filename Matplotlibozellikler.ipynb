```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi2 = numpyDizisi1 ** 2
```

```python
numpyDizisi1
```

```python
(benimFigur ,benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2,color='#009B00')
benimEksen.plot(numpyDizisi2,numpyDizisi1,color='#595CCE')

plt.show()
```

```python
(yeniFigure,yeniEksen) = plt.subplots()
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 2,label = 'kalin cizgi' , color='blue',linewidth = 4.0)
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 5,label = 'noktali cizgi', color='red', linestyle = '-.')
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 9,label = 'cizgi', color='green', linestyle = '--')
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 7,label ='markerli', color='#000000', linestyle = '--' , marker ='o', markerfacecolor='red')
plt.legend()
plt.show()
```
