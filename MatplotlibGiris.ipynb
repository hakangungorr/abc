```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasListesi = [10,20,30,40,50,50,50,60,70,75]
```

```python
kiloListesi = [20,60,63,88,93,76,82,87,99,100]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
numpyKiloListesi
```

```python
numpyYasListesi
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,'r') #(x,y,color)
plt.xlabel('Yas')
plt.ylabel('Kilo')
plt.title('Kilonun yas gore degisim grafigi')
plt.show()
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,'r*-')
plt.show()
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,'r--')
plt.xlabel('Day')  # X-axis label
plt.ylabel('Km')   # Y-axis label
plt.title('Days vs Kilometers')  # Title for the plot
plt.show()
```

```python
plt.subplot(2,1,1) #2sira 1 kolon 1.grafik
plt.plot(numpyDizisi1,numpyDizisi2, 'r*-')
plt.subplot(2,1,2) #2sira 1 kolon 2.grafik
plt.plot(yasListesi,kiloListesi,'r--')
plt.show()
```

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.3,0.3,0.7,0.7])
figureAxes.plot(numpyDizisi1,numpyDizisi2, 'g')
figureAxes.set_xlabel('X ekseni veri ismi')
figureAxes.set_ylabel('Y ekseni veri ismi')
figureAxes.set_title('Grafik Basligi')
plt.show()
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2 = figure2.add_axes([0.3,0.5,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,'g')
eksen1.set_xlabel('x ekseni')
eksen1.set_ylabel('y ekseni')
eksen1.set_title('Ana Grafik Baslik')


eksen2.plot(numpyDizisi2,numpyDizisi1,'r')
eksen2.set_xlabel('x ekseni kucuk')
eksen2.set_ylabel('y ekseni kucuk')
eksen2.set_title('Kucuk grafik')

plt.show()
```

```python
(benimFigure,benimEksen) = plt.subplots()
plt.show()
```

```python
(benimFigure,benimEksenler) = plt.subplots(nrows=1,ncols=2)
type(benimEksen)

for eksen in benimEksenler:
  eksen.plot(numpyDizisi1,numpyDizisi2,'g')
  eksen.set_xlabel('x ekseni')
plt.tight_layout() #goruntuyu ayarlio 
plt.show()
```

```python
yeniFigure = plt.figure() ##dpi=100 

yeniEksen = yeniFigure.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi2,label = 'Numpy dizisi ** 2')
yeniEksen.plot(numpyDizisi1,numpyDizisi1**4, label = 'Numpy dizisi ** 4')
yeniEksen.legend(loc=2)
plt.show()
```
